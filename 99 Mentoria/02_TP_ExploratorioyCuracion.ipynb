# 2. Etiquetas de variables/columnas: no usar caracteres especiales

Chequar que no haya caracteres fuera de `a-Z`, `0-9` y `_` en los nombres de columnas de los Dataframes:

```python
# TODO

#caracteres especiales...
import re

def check(col):
    regex = re.match(r'([a-zA-Z0-9_])+$', col)
     
    if regex == None:
        validacion = 'INVALIDO'
    else:
        validacion = 'ok'
    
    return validacion

#side by side...
from IPython.display import display_html

def siamesas(*args):
    html_str = ''
    spaciador = '<table style="min-width: 40px !important;"><tr style="min-width: 40px !important; background:none !important;"><td style="min-width: 40px !important;"></td></tr></table>'
    for df in args:
        html_str += df.to_html() + spaciador
        
    display_html(html_str.replace('table', 'table style = "display:inline"'), raw = True)
```

```python
#validamos nombres de columnas...

dfs = []
cols = []
estados = []

for df in ['player_df', 'team_df', 'match_df']:
    for col in pd.eval(df + '.columns'):
        #print(df, col, '->', check(col))
        dfs.append(df)
        cols.append(col)
        estados.append(check(col))

dfs = pd.DataFrame.from_dict({'tabla':dfs, 'columna':cols, 'valida':estados})
siamesas(dfs.loc[dfs.tabla == 'player_df'], dfs.loc[dfs.tabla == 'team_df'], dfs.loc[dfs.tabla == 'match_df'])
```

### Imputacion usando Media y Moda

```python
#aplicamos esto al ejemplo anterior: Sevilla FC vs Real Sporting de Gijón...

match_df.loc[(match_df['home_team_long_name'] == 'Sevilla FC') & (match_df['away_team_long_name'] == 'Real Sporting de Gijón')].fillna(match_df.loc[(match_df['home_team_long_name'] == 'Sevilla FC') & (match_df['away_team_long_name'] == 'Real Sporting de Gijón')].mean())
```

```python
# y ahora intentamos generalizarlo a todas las ligas menos 'Switzerland Super League' y 'Poland Ekstraklasa...

NaNs = match_df['B365H'].isnull().groupby(match_df['league_name']).agg({'NaN':sum}).sort_values('NaN', ascending = True).reset_index()
NaNs = NaNs.loc[NaNs.NaN > 0]
NaNs
```

```python
#ejecutamos sobre todas las ligas menos 'Switzerland Super League' y 'Poland Ekstraklasa...

for i, r in NaNs.iterrows():
    #print(i, r['league_name'])
    display(match_df.loc[(match_df['league_name'] == r['league_name']) & (match_df['B365H'].isnull())])
    
    juegos_nan = match_df.loc[(match_df['league_name'] == r['league_name']) & (match_df['B365H'].isnull())]
    
    c = 0
    for i1, r1 in juegos_nan.iterrows():
        local_nan = juegos_nan.iloc[c].home_team_long_name
        visita_nan = juegos_nan.iloc[c].away_team_long_name
        display(match_df.loc[(match_df['home_team_long_name'] == local_nan) & (match_df['away_team_long_name'] == visita_nan)])
        display(match_df.loc[(match_df['home_team_long_name'] == local_nan) & (match_df['away_team_long_name'] == visita_nan)].fillna(match_df.loc[(match_df['home_team_long_name'] == local_nan) & (match_df['away_team_long_name'] == visita_nan)].mean()))
        c = c + 1
        idx = match_df.loc[(match_df['home_team_long_name'] == local_nan) & (match_df['away_team_long_name'] == visita_nan)].index
        d = match_df.loc[idx, ['B365H', 'B365D', 'B365A']].mean().to_dict()
        match_df.loc[idx, ['B365H', 'B365D', 'B365A']] = match_df.loc[idx, ['B365H', 'B365D', 'B365A']].fillna(d)
    
    if i > 7:
        break
```

# 5. Normalizacion de columnas

Primero que todo la notación:
* $x = [x_1, x_2, ..., x_n]$
* $\mu$: Media
* $\sigma$: Desviacion Estandar

Ahora normalizaremos algunas de las columnas del Dataframe, para ello usaremos dos tipos de normalizacion:

* Min-Max:  
$$
z_i = \frac{x_i - min(x)}{max(x) - min(x)}
$$
* Z-score
$$
z_i = \frac{x_i - \mu}{\sigma}
$$

Normalizar la columna `crossing` usando **Min-Max**.

Normalizar la columna `short_passing` usando **Z-score**.

**Hints**:
* https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html#sklearn.preprocessing.MinMaxScaler
* https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.zscore.html

### Min-Max

```python
# TODO
from sklearn.preprocessing import MinMaxScaler

pre = player_df['crossing'].head().copy()
minmaxscaler = MinMaxScaler()
player_df['crossing'] = minmaxscaler.fit_transform(player_df['crossing'].values.reshape(-1, 1))
pos = player_df['crossing'].head().copy()

siamesas(pd.DataFrame(pre).rename(columns={'crossing': 'crossing(pre)'}), pd.DataFrame(pos).rename(columns={'crossing': 'crossing(pos)'}))
```

### Z-score

```python
# TODO
from scipy import stats

pre = player_df['short_passing'].head().copy()
player_df['short_passing'] = stats.zscore(player_df['short_passing'])
pos = player_df['short_passing'].head().copy()

siamesas(pd.DataFrame(pre).rename(columns={'short_passing': 'short_passing(pre)'}), pd.DataFrame(pos).rename(columns={'short_passing': 'short_passing(pos)'}))
```
