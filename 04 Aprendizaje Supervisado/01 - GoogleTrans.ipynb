#### El plan es recorrer el campo 'Description' y traducir a english aquellos textos que están en otros idiomas para luego aplicarles alguna técnica de PLN.

```python
#librerias...

import numpy as np
import pandas as pd
import re
```

```python
#levantamos train y test originales...

train_df = pd.read_csv('data/train.csv')
test_df = pd.read_csv('data/test.csv')

display(train_df.head(), test_df.head())
```

```python
#para remover emojis y rarezas en 'Description'...
import re

emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
```

```python
#language Translator...
from googletrans import Translator
import time
```

```python
#identificamos 'Description' NO 'english' y 'horrores' en train_df...

for index, row in train_df.iterrows():
    translator = Translator() #objetotTranslator.
    
    try:
        descrip = emoji_pattern.sub(r'', row['Description'])
        translated_train_df = translator.translate(descrip)
        
        if translated_train_df.src != 'en':
            #print('original:', row['Description'])
            #print('idioma origen:', translated_train_df.src)
            #print('idioma destino:', translated_train_df.dest)
            #print('traducción:', translated_train_df.text)
            train_df.loc[index, 'lang'] = translated_train_df.src
            print('Traduciendo fila:{}'.format(index))
            train_df.loc[index, 'Description'] = translated_train_df.text
        else:
            train_df.loc[index, 'lang'] = 'en'
    except:
        print('horror: ({}) {}'.format(index, row['Description']))
        train_df.loc[index, 'lang'] = 'xx'
        train_df.loc[index, 'Description'] = np.nan
        
    time.sleep(2)
    #input('pausa')
```

```python
#limpiamos train_df...
train_df.loc[train_df['lang'] == 'xx', 'Description'] = np.nan

#generamos 'Description_clean' sin emojis...
train_df['Description_clean'] = train_df['Description'].str.replace('[^\w\s#@/:%.,_-]', '', flags = re.UNICODE)

#... backupeamos en disco.
train_df.to_csv('data/train_en.csv', header = True, index_label = 'ind')
```

```python
#identificamos 'Description' NO 'english' y 'horrores' en test_df...

for index, row in test_df.iterrows():
    translator = Translator() #objetotTranslator.
    
    try:
        descrip = emoji_pattern.sub(r'', row['Description'])
        translated_test_df = translator.translate(descrip)
        
        if translated_test_df.src != 'en':
            #print('original:', row['Description'])
            #print('idioma origen:', translated_test_df.src)
            #print('idioma destino:', translated_test_df.dest)
            #print('traducción:', translated_test_df.text)
            test_df.loc[index, 'lang'] = translated_test_df.src
            print('Traduciendo fila:{}'.format(index))
            test_df.loc[index, 'Description'] = translated_test_df.text
        else:
            test_df.loc[index, 'lang'] = 'en'
    except:
        print('horror: ({}) {}'.format(index, row['Description']))
        test_df.loc[index, 'lang'] = 'xx'
        test_df.loc[index, 'Description'] = np.nan
        
    time.sleep(2)
    #input('pausa')
```
